## Graphs

### *Histogram*

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
age = df.Age
ax.hist(age, 30)

ax.set_title('Age distribution in FIFA 2019')
ax.set_xlabel('Age')
ax.set_ylabel('Number of players')
```

The histogram visually confirms that the mean age of a FIFA player is between 22 and 23. There also seems to be a reasonably linear decrease in players as they get older, which isn't terribly surprising. The decrease appears roughly linear until around 35; based on the slope I estimate that roughly 120 players above the age of 28 retire each year.

### *Correlation matrix*

```python
corr = df.corr()
corr
```

## *Heat map*

```python
plt.figure(figsize=(40,24))
sns.heatmap(corr, annot=True, annot_kws={'size':13})
```

* Unsurprisingly, there's a 0.89 correlation between shot power and long shots
* 0.92 correlation between acceleration and sprint speed - fast people run fast, I guess
* 0.84 correlation between finishing and penalties. Seems like the people who go for goal shots also tend to get more penalties, though it's interesting that there's only a 0.24 correlation between agression and finishing and 0.34 between aggression and penalties; I would've expected those correlations to be stronger.

* Strength and balance are negatively correlated at -0.39, which makes sense as the weaker you are the more likely you are to get pushed over.
* ID also has a couple strong negative correlations which is strange until we note that it's also strongly negatively correlated with age. So, the older you are the lower your number. And the lower your number, the better your international reputaion (-0.36), reactions (-0.41), and composure (-0.38), all of which paints a picture of older players being more responsive, better respected, and more composed.


## *Line chart* 

```python
fig, ax1 = plt.subplots()

ax1.set_xlabel('Age (years)')
ax1.set_ylabel('Reactions & Composure (Percentile)', color='purple')
lns1 = ax1.plot(df.groupby('Age')['Reactions'].agg('mean'), label='Reactions', color='red')
lns2 = ax1.plot(df.groupby('Age')['Composure'].agg('mean'), label='Composure', color='purple')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()

color = 'green'
ax2.set_ylabel('International Reputation', color=color)
lns3 = ax2.plot(df.groupby('Age')['International Reputation'].agg('mean'), color=color, label='International Reputation')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()

lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs)

plt.title('Older players are more well-known and have better reactions and composure')

plt.show()
```

## *Scatter plot*

```python
plt.scatter(x=df.Finishing, y=df.ShotPower, s=1)
plt.xlabel('Finishing')
plt.ylabel('Shot Power')
plt.title('Stronger shooters take more shots on the goal')
plt.show()
```
